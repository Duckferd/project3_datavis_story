```python
from pandas import json_normalize
import json
from pprint import pprint
import pandas as pd
```

```python
path = "./mygeodata/CMHC_ZONE.geojson"
with open(path) as user_file:
    parsed_json = json.load(user_file)
```

```python
vancouverzones = pd.read_csv("./2022table.csv",
                             usecols = [0],
                             skiprows=1,
                             header=1,
                             names = ["CMHC_Zone"])
print(vancouverzones)
```

```python
cleanvanzones = vancouverzones['CMHC_Zone']
```

```python
cleanvanlist = []
for i in range(0,26):
    cleanvanlist.append(cleanvanzones[i])

print(cleanvanlist)
```

```python
# Print a document to examine JSON structure and count documents in raw geoJSON
pprint(parsed_json['features'][279])
doccount = len(parsed_json['features'])
print(f'There are {doccount} records in the original geoJSON file from CMHC.')
```

```python
cleaned_JSON = []
index_list_JSON = []
for i in range(doccount):
    if parsed_json['features'][i]['properties']['ZONENAME_E'] in cleanvanlist:
        cleaned_JSON.append(parsed_json['features'][i])
        index_list_JSON.append(i)
pprint(cleaned_JSON)
print(index_list_JSON)
```
