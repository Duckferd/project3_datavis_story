```python
import pandas as pd
import glob as glob
import os
import re
```

```python
# Get Data file names 
path = '../Raw_Data/CMHC Rent Data CSVs - 2010 to 2017/'
filenames = glob.glob(os.path.join(path + "/*.csv"))

dfs = list()
for f in filenames:
    data = pd.read_csv(f, encoding= "unicode_escape", header=2)
    data.drop(data.index[26:34], axis=0, inplace=True)
    data.drop(columns=['Unnamed: 2', 'Unnamed: 4', 'Unnamed: 6', 'Unnamed: 8', 'Unnamed: 10', 'Unnamed: 12'], inplace=True)
    # use split method to extract 'Year' column from the datafile path
    data['Year'] = f.split("\\")[-1].split(' Data.')[0]
    dfs.append(data)

rental_df = pd.concat(dfs, ignore_index=True)

```

```python
# Rename Neighbbourhood column
rental_df_clean1 = rental_df.rename(columns={'Unnamed: 0': 'Neighbourhood'})

rental_df_clean1
```

```python
rental_df_clean = rental_df_clean1.replace(['++'], '0')

rental_df_clean['Neighbourhood'] = rental_df_clean['Neighbourhood'].astype(str)
rental_df_clean['Units'] = rental_df_clean["Units"].replace("[,]", "", regex=True).astype(int)
rental_df_clean['Average Rent ($)'] = rental_df_clean["Average Rent ($)"].replace("[,]", "", regex=True).astype(int)
rental_df_clean['Median Rent ($)'] = rental_df_clean["Median Rent ($)"].replace("[,]", "", regex=True).astype(int)
rental_df_clean['Vacancy Rate (%)'] = rental_df_clean['Vacancy Rate (%)'].astype(float)
rental_df_clean['% Change'] = rental_df_clean['% Change'].astype(float)
rental_df_clean['Year'] = rental_df_clean['Year'].astype(int)

rental_df_clean.dtypes
```

```python
rental_df_clean.to_csv('Rental_data_clean.csv', index=False, header=True)
```
