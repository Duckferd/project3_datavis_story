```python
import pandas as pd
```

```python
file = "../Raw_Data/crime.csv"
```

```python
df = pd.read_csv(file)
df.head()
```

```python
columns=['TYPE','YEAR','MONTH', 'Latitude', 'Longitude']
```

```python
reduced_df=df.loc[:,columns]
reduced_df.head()
```

```python
reduced_df.loc[df['TYPE'] == "Break and Enter Residential/Other"]
```

```python
df2=reduced_df.replace( "Break and Enter Residential/Other", "Break and Enter Residential")
```

```python
df2
```

```python
final_df=df2.loc[(reduced_df['YEAR']>=2010) & (reduced_df['YEAR']<=2017) & (reduced_df['Latitude']>0 & (reduced_df['Longitude']>0))]
len(final_df)
```

```python
final_df.to_csv("crime_data_clean.csv", encoding="utf-8", index=False, header=True)
```
